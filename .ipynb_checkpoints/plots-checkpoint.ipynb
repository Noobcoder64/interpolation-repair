```python
import matplotlib.pyplot as plt
```

```python
# Sample data
categories = ['interpolation-new', 'interpolation', 'GLASS', 'JVTS-Repair']
values = [4, 9, 9, 0]

# Create a bar chart
bars = plt.bar(categories, values)
plt.ylim(0, 9)

# Add labels and title
plt.title('SYNTECH15-UNREAL')
plt.ylabel('# Repaired')

# Add labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 2), ha='center', va='bottom')

# Show the chart
plt.show()
```
